Se decidió elaborar el algoritmo de solución para cualquier número de trapezoides (<i>n</i>) en lenguaje de programación Python.

La fórmula para este fin está definida en la página 10 del referido texto (ecuación 1.16) de la siguiente forma: 

$$\int_a^b f(x) \,dx \approx  h[\frac{1}{2} f(x_{0}) + \sum_{i=1}^{n-1} f(x_{i}) + \frac{1}{2} f(x_{n})]$$  

En cuanto al error relativo definido para el método trapezoidal, también se utiliza un algoritmo en Python basado en la ecuación propuesta en el libro de texto como se indica a continuación:

$$\frac{|u-v|}{|u|}*100\%$$

Donde <b><i>u</i></b> es el resultado de la integral exacta y <b><i>v</i></b> es la aproximación generada por el método trapezoidal que estamos utilizando.

```python
# Para obtener la representación gráfica del área bajo la curva

# se procede a definir el rango de x y la escala.
x = np.linspace(-3,3, 40) # secuencia númerica de x entre -3 y 3 con 40 saltos totales para mejor visualización.
x_int = np.linspace (0,1) # sabemos que el área bajo la curva va en este rango de integración

# Se define la función original y y la función donde sólo corre el rango de integración.
y = (1/(1+x))
y_int=(1/(1+x_int))

#Se crea la gráfica

ax = plt.axes() # se define el objeto con las axisas
plt.plot(x,y)
plt.grid()
plt.fill_between(x_int,y_int, color='C3')
plt.title('Área bajo la curva\nSolución Exacta')

# Apariencia de la gráfica:

ax.set_facecolor("Azure")
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_position('zero')
ax.spines['right'].set_position('zero')

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Se define la función a integrar

def f1(x):
    return 1/(1+x)

# El número de trapezoides que nos interesan
n = 3

# Se prepara la gráfica
fig = plt.figure()
ax = fig.add_subplot(111)#(322)


# Nuevamente el rango de la función con la idea de que sea más claro el ejemplo
x = np.linspace(-0.1,1.1, 50)
xtrap = np.linspace(0, 1, n+1) # Se debe sumar n por 1 porque la función linspace debe tomar un parámetro de item > 2 para empezar a operar.

# Se dibuja el área bajo la curva
ax.fill_between(xtrap, 0, f1(xtrap), facecolor='red',edgecolor='darkred', linewidth=2)

# La función a integrar
ax.plot(x, f1(x),'blue', linestyle='--')  # Function

# La función para dibujar los bordes de los polígonos.
plt.stem(xtrap,f1(xtrap), linefmt="black")


# Apariencia
plt.grid()
plt.title('Área bajo la curva\nSolución Trapezoidal')
ax.set_facecolor("Azure")
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_position('zero')
ax.spines['right'].set_position('zero')

plt.show()
```
